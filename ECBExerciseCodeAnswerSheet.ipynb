#### question 1

```python
# change date data from string type to a datetime like object
dataframe['ViolationDate'] = pd.to_datetime(dataframe['ViolationDate'])
dataframe['DueDate'] = pd.to_datetime(dataframe['DueDate'])
dataframe['LastPaidDate'] = pd.to_datetime(dataframe['LastPaidDate'])
dataframe['HearingDate'] = pd.to_datetime(dataframe['HearingDate'])
dataframe['HearingRequestReceivedDate'] = pd.to_datetime(dataframe['HearingRequestReceivedDate'])
```

```python
# index the dataframe by the date of citation issuance
dataframe.set_index('ViolationDate',inplace=True)
```

```python
print(dataframe.index.min())
print(dataframe.index.max())

print('This dataset covers a period of %s' %((dataframe.index.max() - \
                                              dataframe.index.min()).days) + ' days.')
```

#### question 7

```python
print('Total number of unique ECB officer ids is %s.' %(dataframe['OfficerID'].nunique()))
```

```python
#number of unique officer IDs in each year
dataframe.resample('A')['OfficerID'].nunique()
```

### <font color=green>Taking a quick peak.</font>

```python
fig,ax1 = plt.subplots()
dataframe.resample('A')['OfficerID'].nunique().plot(lw=4,color='blue')
plt.ylabel('number of officer IDs in field')
plt.xlabel('Year of Issuance')
ax1.yaxis.label.set_color('green')

ax2 = ax1.twinx()
dataframe.resample('A')['CitationNo'].count().plot(ls='--',color='green',\
                                                   alpha=.7)
ax2.yaxis.label.set_color('blue')
plt.ylabel('number of citations issued')
plt.title('ECB Inspector Workforce & Citation Volume')
plt.grid(False)

```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(1,2,1)
plt.scatter(dataframe.resample('A')['OfficerID'].nunique(),\
           dataframe.resample('A')['CitationNo'].count())
plt.title('Inspector & Citations Volume')
plt.ylabel('number of ECB citations')
plt.xlabel('number of uniquie Officer IDs')
ax1.yaxis.set_label_position('right')


### EXPERIMENTING ON SOMETHING WITH SECOND SUBPLOT####
test = pd.DataFrame(dataframe.groupby('OfficerID')['CitationNo'].count())
test.reset_index(inplace=True)
######################################################

ax2 = fig.add_subplot(1,2,2)
plt.scatter(test.index,test['CitationNo'],color='red',alpha=.1,edgecolor='blue')
plt.title('Quick Glance by Officer')
plt.xlabel('annoymous officer number');
```
